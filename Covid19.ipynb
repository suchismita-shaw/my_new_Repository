```python
import pandas as pd
```

```python
data = pd.read_csv ("/content/Latest Covid-19 India Status.csv")
```

```python
data.columns
```

# Data Cleaning

```python

data.drop_duplicates(inplace=True)
```

```python
data
```

```python
print("\nMissing Values in Each Column:")
print(data.isnull().sum())
```

```python
numeric_cols = ['Total Cases', 'Active', 'Discharged', 'Deaths',
                'Active Ratio', 'Discharge Ratio', 'Death Ratio', 'Population']
data[numeric_cols] = data[numeric_cols].fillna(data[numeric_cols].median())
```

```python
data.head()
```

```python
high_case_states = data[data['Total Cases'] > 10000]
print(f"\nStates/UTs with more than 10,000 total cases:")
print(high_case_states[['State/UTs', 'Total Cases']])
```

```python
agg_results = data.agg({
    'Total Cases': ['mean', 'sum'],
    'Active': ['mean', 'sum'],
    'Discharged': ['mean', 'sum'],
    'Deaths': ['mean', 'sum']
})
print("\nAggregation Results (Mean and Sum):")
print(agg_results)
```

```python
data['Case Fatality Rate (%)'] = (data['Deaths'] / data['Total Cases']) * 100
```

```python
data['Cases per Million'] = (data['Total Cases'] / data['Population']) * 1_000_000
```

```python
processed_file_path = 'processed_covid19_dataset.csv'
data.to_csv(processed_file_path, index=False)
print(f"\nProcessed dataset saved to: {processed_file_path}")
```

```python
summary_stats = data.describe()
print("\nSummary Statistics:")
print(summary_stats)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
plt.figure(figsize=(10, 6))
plt.bar(data['State/UTs'], data['Total Cases'], label='Total Cases', color='blue', alpha=0.7)
plt.bar(data['State/UTs'], data['Deaths'], label='Deaths', color='red', alpha=0.7)
plt.xlabel('State/UTs')
plt.ylabel('Count')
plt.title('Comparison of Total Cases and Deaths by State/UT')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
fig = px.scatter(data, x='Active', y='Discharged',
                 size='Total Cases', color='State/UTs',
                 hover_name='State/UTs', title='Active vs. Discharged Cases')
fig.show()
```

```python
correlation_matrix = data[['Total Cases', 'Active', 'Discharged', 'Deaths',
                           'Active Ratio', 'Discharge Ratio', 'Death Ratio', 'Population']].corr()
```

```python
correlation_matrix
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of COVID-19 Statistics')
plt.show()
```
